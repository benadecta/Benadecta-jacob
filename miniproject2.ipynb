```python

import ipywidgets as widgets
from IPython.display import display, HTML

# Create a shopping list widget
shopping_list = []

# Style
style = {'description_width': 'initial'}

# Center the "Enter Item" label and text input field using HTML and CSS
item_label = widgets.HTML('<label style="text-align: center;">Enter Item:</label>')
item_entry = widgets.Text(layout=widgets.Layout(width="90%"))

add_button = widgets.Button(description="Add Item", style=style, button_style='success')
remove_button = widgets.Button(description="Remove Item", style=style, button_style='danger')
display_button = widgets.Button(description="Display List", style=style, button_style='info')
save_button = widgets.Button(description="Save List to File", style=style, button_style='warning')
output = widgets.Output()

def add_item(_):
    item = item_entry.value
    if item:
        shopping_list.append(item)
        item_entry.value = ""

def remove_item(_):
    item = item_entry.value
    if item in shopping_list:
        shopping_list.remove(item)
        item_entry.value = ""

def display_list(_):
    with output:
        print("Shopping List:")
        for item in shopping_list:
            print(f"- {item}")

def save_list(_):
    with open("shopping_list.txt", "w") as file:
        for item in shopping_list:
            file.write(item + "\n")

add_button.on_click(add_item)
remove_button.on_click(remove_item)
display_button.on_click(display_list)
save_button.on_click(save_list)

# Create the GUI layout
button_box = widgets.HBox([add_button, remove_button, display_button, save_button])
layout = widgets.VBox([
    item_label,
    item_entry,
    button_box,
    output
])

# Apply some additional styling to the layout
layout.layout.margin = 'auto'  # Center the layout
layout.layout.max_width = '500px'  # Limit the width to make it visually appealing

# Center the input field text using HTML and CSS
display(HTML("<style>.widget-container { text-align: center; }</style>"))
display(layout)
```
